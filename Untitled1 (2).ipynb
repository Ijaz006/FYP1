```python
from ultralytics import YOLO
import cv2
from google.colab import files
import numpy as np

# Upload the video from your Mac
uploaded = files.upload()

# Get the uploaded video file name
video_path = list(uploaded.keys())[0]



# Open the video file
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create a VideoWriter object
out = cv2.VideoWriter("output_video.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (frame_width, frame_height))

# Process each frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame from BGR to RGB (YOLO expects RGB format)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)


    # Run inference on the frame
    results = model1(frame_rgb)

    # Annotate the frame with detection results
    annotated_frame = results[0].plot()

    # Convert back to BGR for OpenCV compatibility
    annotated_frame_bgr = cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR)

    # Write the frame with detections to the output video
    out.write(annotated_frame_bgr)

# Release video objects
cap.release()
out.release()

# Download the output video
files.download("output_video.mp4")
```

```python
import cv2
import numpy as np
import pytesseract
from PIL import Image
import supervision as sv
from google.colab import files
import matplotlib.pyplot as plt  # To display images in Colab

# Upload image from PC
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Get uploaded file name

# Load the image using OpenCV
image = cv2.imread(image_path)

# Display Original Image
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()

# Run YOLO model inference
results = model.predict(image, confidence=50, overlap=30).json()

# Convert results to Supervision Detections
detections = sv.Detections.from_inference(results)

# Ensure at least one number plate is detected
if len(detections.xyxy) > 0:
    # Get bounding box of first detected number plate
    x_min, y_min, x_max, y_max = map(int, detections.xyxy[0])

    # Crop the detected number plate from the image
    plate_image = image[y_min:y_max, x_min:x_max]

    # Convert to grayscale (since you want it in grayscale)
    plate_gray = cv2.cvtColor(plate_image, cv2.COLOR_BGR2GRAY)

    # Apply Adaptive Thresholding
    plate_thresh = cv2.adaptiveThreshold(
        plate_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )

    # Apply Morphological Operations to remove noise
    kernel = np.ones((1, 1), np.uint8)
    plate_thresh = cv2.morphologyEx(plate_thresh, cv2.MORPH_CLOSE, kernel)

    # Resize for better OCR recognition
    plate_thresh = cv2.resize(plate_thresh, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

    # Save processed plate image (Grayscale)
    cv2.imwrite("processed_plate.jpg", plate_thresh)

    # Display the Extracted Number Plate (Grayscale)
    plt.figure(figsize=(6, 4))
    plt.imshow(plate_thresh, cmap="gray")
    plt.title("Extracted Number Plate (Grayscale)")
    plt.axis("off")
    plt.show()

    # Apply OCR to extract both letters and numbers
    plate_text = pytesseract.image_to_string(
        plate_thresh, config="--psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
    )

    # Clean OCR output (remove unwanted characters)
    plate_number = "".join(filter(str.isalnum, plate_text))

    print(f"‚úÖ Extracted Number Plate: {plate_number}")

else:
    print("‚ùå No number plate detected!")
```
